```python
#Q5) The following code uses stub functions to design the game. Complete the code.

import random

def generate_random_num():
    """
    Generates a random integer between 1 and 100 (inclusive) for the user to guess.
    """
    return random.randint(1, 100)


def get_user_input():
    """
    This function asks the user to guess the number.
    Makes sure the number entered is an integer.
    """
    while True:
        try:
            guess_num = int(input("Guess a number between 1 and 100: "))
            return guess_num
        except ValueError:
            print("That's not a valid integer. Please try again.")


def check_the_number(actual_num, guess_num):
    """
    Checks if the actual number and guessed number are the same.
    """
    return actual_num == guess_num


def display_the_feedback(is_correct, chances):
    """
    Prints feedback based on whether the guess is correct or not.
    """
    if is_correct:
        print("üéâ Bingo! You guessed the correct number.")
    else:
        print(f"‚ùå Wrong guess. Chances left: {chances}")


### Do not change the functions below ###
def guess_num_game():
    actual_num = generate_random_num()

    max_chances = 10
    for counter in range(1, max_chances + 1):
        guess_num = get_user_input()
        is_correct = check_the_number(actual_num, guess_num)
        display_the_feedback(is_correct, max_chances - counter)
        if is_correct:
            return "You did it! You won the game"

    return "You lost! Play one more time?"


def main():
    while True:
        win_lost_msg = guess_num_game()
        print(win_lost_msg)
        is_continue = int(input("Enter 1 to continue: "))
        if not is_continue:
            return "Bye. Thanks for playing."


main()
```
